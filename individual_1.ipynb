<p>Постройте график зависимости скорости от времени, если тело в момент времени t = 2с со скоростью u = 24м/с. Движение равноускоренное, начальная скорость равна u<sub>0</sub> = 13м/c. Определить по графику путь за 30 секунд.</p> 

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
u_start = 13
u_1 = 24 
t_given = 2
a = (u_1 - u_start ) / t_given
t = np.array([i for i in range(31)])
u = np.array([u_start + a * i for i in t])
```

```python
plt.title('Зависимость скорости от пути')
plt.xlabel("Скорость")
plt.ylabel("Время")
plt.grid(True)

plt.plot(t, u, 'p-g', label="first", lw=5, mec='b', mew=1, ms=10, mfc='b')
plt.fill_between(t,u)
```
