<p>Статистика трат за последние 4 месяца (Статистика мп Тинькофф)</p>

```python
import matplotlib.pyplot as plt
%matplotlib inline

fruits = ["Февраль","Март", "Апрель", "Май",]
counts = [22990.80, 31857.10, 44477.33, 20552.19]
plt.bar(fruits, counts)
plt.title("Траты за 4 месяца")
plt.xlabel("Месяц")
plt.ylabel("Кол-во потраченных средств")
```
